```python
def batch_tuple(n, batch_size):
    
    remainder = n % batch_size
    if remainder == 0:
        it = n//batch_size
    else:
        it = n//batch_size + 1
    
    start = 0
    li = []
    for i in range(it):
    
        if i == it-1:
            end = start + remainder
            tupl = (start,end)
            li.append(tupl)
            continue
        
        end = (i+1)*batch_size
        tupl = (start,end)
        li.append(tupl)
        start = end
        
    return(li)
```

```python
def remove_extra_line(text_list):
    new_list = []
    for item in text_list:
        if item == '':
            continue
        new_list.append(item)
    return(new_list)
```

```python
def text_slice(file_name,batch_size=10, is_remove_space = False):
    x = open(file_name, 'r')    
    all_lines = [item.strip() for item in x.readlines()]
    if is_remove_space == True:
        all_lines = remove_extra_line(all_lines)
    n = len(all_lines)

    batch_result = batch_tuple(n,batch_size)
    list_lines = []
    for item in batch_result:
        start = item[0]
        end = item[1]
        a = tuple(all_lines[start:end])
        list_lines.append(a)
    return(list_lines)
    
            
```

```python
result = text_slice('sample.txt', is_remove_space=True)
```
