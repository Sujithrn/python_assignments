```python
x = open('sample.txt','r')
content = x.read()
y = x.readlines()
```

```python
list_lines = []
for item in y:
    list_lines.append([item.strip('\n')])
```

```python
content
```

```python
split_lines = []
for item in list_lines:
    for line in item:
        line = line.split()
        split_lines.append(line)
```

```python
split_lines
```

```python
words = content.split()
b = {}
i = 0

for word in words:
    if word in b:
        continue
    else:
        b[word] = i
    i = i+1
        
```

```python
b
```

```python
def map_words_index(word_list, dict_index):
    temp2 = []
    for w in word_list:
        temp2.append(dict_index[w])
    return temp2
```

```python
sample = ['this', 'is', 'good']
```

```python
for word in sample:
    print(word,b[word])
```

```python
temp = []
for item in split_lines:
    temp2=map_words_index(item, b)
    temp.append(temp2)
        
        
    
```
